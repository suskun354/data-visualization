```python
#Özellkik mühendsiliği
df["Date"]
```

```python
df["Date"].dtype #tarih olarak alıyor mu kontrol
```

```python
df["Date"]=pd.to_datetime(df["Date"]) #veri türünü değişitirme
df["Date"].info()
```

```python
df["day"]=df.Date.dt.day #tarih içerisnedeki gün olarak ayırır
df.day
```

```python
df["month"]=df.Date.dt.month
df.month
```

```python
df["year"]=df.Date.dt.year
df.year
```

```python
df[["day","month","year"]]
```

```python
df["dayofweek"]=df.Date.dt.dayofweek
df["dayofweek"]
```

```python
df["dayofweek"].unique()
```

```python
df["dayofweek"].replace({0:"Monday", 1:"Tuesday",2:"Wednesday",
                        3:"Thursday", 4:"Friday", 5:"Saturday",
                        6:"Sunday"}, inplace=True) #tarrih gün formatına çevirme
df["dayofweek"]
```

```python
df["day_name"]=df.Date.dt.day_name()
df["day_name"]
```

```python
df["month_name"]=df.Date.dt.month_name()
df["month_name"]
```

```python
df["Time"], df["Time"].dtype
```

```python
df["Time"]=pd.to_datetime(df["Time"])
```

```python
df["Time"].info()
```

```python
df['hour']=df.Time.dt.hour
df['hour']
```

```python
plt.figure(figsize=(12,6)) #grafiğin boyutları
sns.countplot(x=df["month_name"]) #belirlenen kategorilerde toplamı alır
plt.xticks(rotation=90) #etiketi yazısını açısı
plt.show()
```

```python
sns.countplot(x=df["month_name"],
              hue=df["Branch"])
```

```python
sns.countplot(x=df["Gender"])
```

```python
sns.countplot(x=df["Gender"],
              hue=df["Branch"])
```

```python
plt.figure(figsize=(12,6))
plt.title("Cinsiyeti göre aylık alışveriş sayısı")
sns.countplot(x=df["Gender"],
              hue=df["month_name"])
```

```python
plt.figure(figsize=(12,6))
plt.title("Cinsiyeti göre aylık alışveriş sayısı")
sns.countplot(y=df["month_name"],
              hue=df["Gender"])
```
